#### **Ejercicio Figuras Geométricas - POO**

</center>

~~~
El ejercicio consiste en crear 2 Clases Padre (FiguraGeometrica y Color) y 2 Clases Hijas (Cuadrado y Rectángulo), en todas las Clases debemos aplicar Encapsulamiento, Métodos Setter y Getter, y Herencia Múltiple.
~~~

```python
#Clase FiguraGeometrica:
class FiguraGeometrica:
    #Método inicializador para la Clase FiguraGeometrica: 
    def __init__(self, alto, ancho):
        self._alto = alto
        self._ancho = ancho

    #Método str()__ para la Clase FiguraGeometrica:
    def __str__(self):
        return f"Alto: {self._alto} unidades\nAncho: {self._ancho} unidades"

    #Método Getter para el Atributo → alto:
    @property
    def Alto(self):
        return self._alto

    #Método Getter para el Atributo → ancho:
    @property
    def Ancho(self):
        return self._ancho

    #Método Setter para el Atributo → alto:
    @Alto.setter
    def Alto(self, alto):
        self._alto = alto

    #Método Setter para el Atributo → ancho:
    @Ancho.setter
    def Ancho(self, ancho):
        self._ancho = ancho

#---------------------------------------------------------------------------------------------------------------------------

#Clase Color:
class Color:
    #Método inicializador para la Clase FiguraGeometrica: 
    def __init__(self, color):
        self._color = color

    #Método Getter para el Atributo → color:
    @property
    def Color(self):
        return self._color

    #Método Setter para el Atributo → color:
    @Color.setter
    def Color(self, color):
        self._color = color

#---------------------------------------------------------------------------------------------------------------------------

#Clase Cuadrado → Herencia Múltiple de FiguraGeometrica y Color:
class Cuadrado(FiguraGeometrica, Color):
    def __init__(self, lado, color):
        FiguraGeometrica.__init__(self, lado, lado) #Inicializamos Atributos de la Clase FiguraGeometrica
        #NOTA: 
        # El valor lado sutituye los valores alto y ancho pero para las operaciones 
        # se sigue utilizando el nombre original de la variable.
        Color.__init__(self, color) #Inicializamos Atributo de la Clase Color

    #Método __str()__ para la Clase Cuadrado:
    def __str__(self):
        return f"Cuadrado\n{super().__str__()}\nColor: {self._color}"

    #Método para calcular el Área de un Cuadrado:
    def area(self):
        return f"Área: {self._alto * self._ancho} u²"

#---------------------------------------------------------------------------------------------------------------------------

#Clase Rectangulo → Herencia Múltiple de FiguraGeometrica y Color:
class Rectangulo(FiguraGeometrica, Color):
    def __init__(self, alto, ancho, color):
        FiguraGeometrica.__init__(self, alto, ancho) #Inicializamos Atributos de la Clase FiguraGeometrica
        Color.__init__(self, color) #Inicializamos Atributos de la Clase Color

    #Método __str()__ para la Clase Rectangulo:
    def __str__(self):
        return f"Rectangulo\n{super().__str__()}\nColor: {self._color}"

    #Método para calcular el Área de un Rectangulo:
    def area(self):
        return f"Área: {self._alto * self._ancho} u²"

#---------------------------------------------------------------------------------------------------------------------------

#Pruebas del funcionamiento del Código:
ejm1 = Cuadrado(10, "Rojo")
print(ejm1)
print(ejm1.area())

ejm2 = Rectangulo(15, 17, "Verde")
print(ejm2)
print(ejm2.area())

ejm2.Alto = 12
ejm2.Ancho = 8
ejm2.Color = "Amarillo"
print(ejm2)
print(ejm2.area())
```
